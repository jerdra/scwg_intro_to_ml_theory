```python
from sklearn.datasets import make_blobs
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X,y = make_blobs(n_samples=100, n_features=2, random_state=200, centers=3, cluster_std=2)
```

```python
df = pd.DataFrame(np.c_[X,y])
df.columns = ['x1','x2','y']
df['y'] = df['y'].astype(np.int64)
```

```python
ax = sns.scatterplot(x='x1',y='x2',data=df,label='data')
plt.savefig('../figures/kmeans_data.png',dpi=200)
```

```python
c0_mask = df['y'] == 0
c1_mask = df['y'] == 1
c2_mask = df['y'] == 2
```

```python
ax = sns.scatterplot(x='x1',y='x2',data=df,hue='y')
plt.savefig('../figures/kmeans_data_3clust.png',dpi=200)
```

```python
df.loc[c0_mask | c2_mask, '2-clust'] = 0
df.loc[c1_mask, '2-clust'] = 1
df['2-clust'] = df['2-clust'].astype(np.int64)
```

```python
ax = sns.scatterplot(x='x1',y='x2',data=df,hue='2-clust')
plt.savefig('../figures/kmeans_data_2clust.png',dpi=200)
```

```python
def calculate_distances(X,k):
    dist_mat = np.zeros((X.shape[0],k.shape[0]))
    
    for i in np.arange(0,k.shape[0]):
        
        dist_mat[:,i] = np.linalg.norm(X - k[i,:],axis=1)
    return dist_mat

def assign_centroids(X,K):
    return np.argmin(calculate_distances(X,K),axis=1)

def calculate_new_centroid(X,K,a):
    for i in np.arange(0,K.shape[0]):
        centroid = X[np.where(a == i)[0]].mean(axis=0)
        K[i,:] = centroid
    return K
```

```python
#Define initial cluster positions for K = 3
iteration = 0
K = np.zeros((3,2))
K[:,1] = np.random.uniform(-12.5,3,size=3)
K[:,0] = np.random.uniform(-2,14,size=3)

#Plot initial cluster locations
colors = ['r','g','b']
ax = sns.scatterplot(x='x1',y='x2',data=df)
for i,c in enumerate(colors):
    ax.plot(K[i,0],K[i,1],'{}+'.format(c), label='Centroid {}'.format(i))
    ax.legend()

plt.title('Iteration #{}'.format(iteration))
plt.savefig('../figures/kmeans_initialcentroids.png',dpi=200)
```

```python
#Initial cluster assignment
colors = ['r','g','b']
a = assign_centroids(X,K)
#Plot each cluster
fig, ax = plt.subplots()

for i in np.arange(0,K.shape[0]):
    
    #Get assigned X
    c = colors[i]
    assign_X = X[np.where(a==i)[0]]
    ax.plot(assign_X[:,0],assign_X[:,1],'{}.'.format(c),label='Cluster {}'.format(i))
    ax.plot(K[i,0],K[i,1],'{}+'.format(c), label='Centroid {}'.format(i))
    
    #Connector lines!
    for k in assign_X:
        ax.plot([k[0], K[i,0]], [k[1],K[i,1]], '{}'.format(c), linewidth=0.5)
        
ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$')
ax.set_title('Iteration #{}'.format(iteration))
ax.legend()
plt.savefig('../figures/kmeans_iter_{}.png'.format(iteration),dpi=200)
plt.show()
```

```python
#Assign centroids
iteration += 1
a = assign_centroids(X,K)
K = calculate_new_centroid(X,K,a)
colors = ['r','g','b']

#Plot each cluster
fig, ax = plt.subplots()

for i in np.arange(0,K.shape[0]):
    
    #Get assigned X
    c = colors[i]
    assign_X = X[np.where(a==i)[0]]
    ax.plot(assign_X[:,0],assign_X[:,1],'{}.'.format(c),label='Cluster {}'.format(i))
    ax.plot(K[i,0],K[i,1],'{}+'.format(c), label='Centroid {}'.format(i))
    
    #Connector lines!
    for k in assign_X:
        ax.plot([k[0], K[i,0]], [k[1],K[i,1]], '{}'.format(c), linewidth=0.5)

ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$')
ax.set_title('Iteration #{}'.format(iteration))
ax.legend()
plt.savefig('../figures/kmeans_iter_{}.png'.format(iteration),dpi=200)
plt.show()
```

```python
new_K = calculate_new_centroid(X,K,assign_centroids(X,K))
```

```python
def calculate_mean_sse(X,K,a):
    
    val = 0
    counter = 0
    for i in np.arange(0,K.shape[0]):
        counter += 1
        assign_X = X[np.where(a == i)[0]]
        dists = np.linalg.norm(assign_X - K[i,:],axis=1).mean(axis=0)
        val += dists
    
    return val/float(counter)
```

```python
#Calculate MSE score based on each cluster solution size
tol = 0.0001
mean_sse = []
for k in [1,2,3,4,5]:
    
    #Set up centroids
    K = np.zeros((k,2))
    K[:,1] = np.random.uniform(-12.5,3,size=k)
    K[:,0] = np.random.uniform(-2,14,size=k)
    
    #Loop until convergence
    tol_not_reached = True
    while tol_not_reached:
        
        #Assign and re-calculate centroids
        a = assign_centroids(X,K)
        new_K = calculate_new_centroid(X,K,a)
        
        #Check convergence
        if np.linalg.norm(new_K - K,axis=1).mean() < tol:
            tol_not_reached = False
        K = new_K
            
    #Calculate SSE
    sse = calculate_mean_sse(X,K,a)
    mean_sse.append(sse)
    
        
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(1,6), mean_sse,'k.',markersize=10)
ax.plot(np.arange(1,6),mean_sse,'k')
ax.set_xlabel('K')
ax.set_ylabel('Mean SSE')
ax.set_title('Cluster MSE as a function of Number of Centroids')
plt.savefig('../figures/kmeans_k_vs_mse.png',dpi=200)
```
