```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
X, y = make_regression(n_samples=400, n_features=1, n_informative=1, noise=30,
                      random_state=200, bias=30, n_targets=1)
```

```python
df = pd.DataFrame(np.c_[X,y])
df.columns = ['X','y']
```

```python
train_inds = df.sample(frac=0.60,replace=False).index
test_inds = [d for d in df.index if d not in train_inds]
```

```python
df.loc[train_inds,'holdout'] = 'train'
df.loc[test_inds,'holdout'] = 'test'
```

```python
plt.style.use('seaborn')
```

```python
sns.scatterplot(x='X',y='y',data=df)
plt.savefig('../figures/holdout_reg.png',dpi=200)
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x='X',y='y',data=df,hue='holdout')
plt.savefig('../figures/holdout_split.png',dpi=200)
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x='X',y='y',data=df[df['holdout'] == 'train'])
plt.savefig('../figures/holdout_training.png',dpi=200)
plt.show()
```

Fit the model on training data

```python
trainX = df.loc[df['holdout'] == 'train','X'].values
trainy = df.loc[df['holdout'] == 'train','y'].values
testX = df.loc[df['holdout'] == 'test','X'].values
testy = df.loc[df['holdout'] == 'test','y'].values

trainX = np.c_[np.ones(trainX.shape[0]), trainX]
```

```python
c,r,_,_ = np.linalg.lstsq(trainX,trainy)
```

```python
#Plot training fit
n_train = df[df['holdout'] == 'train'].shape[0]
predX = np.linspace(-5,5,100)
evalX = predX * c[1] + c[0]

ax = sns.scatterplot(x='X',y='y',data=df[df['holdout'] == 'train'], label='training data')
ax.plot(predX,evalX,'k',label='model fit')
ax.text(0.7,0.2,'MSE: {:.2f}'.format(r[0]/n_train),transform=ax.transAxes,fontsize=15)
ax.legend()
plt.savefig('../figures/holdout_trained.png',dpi=200)
plt.show()
```

```python
#Overlay fit over test data and re-compute MSE
n_test = df[df['holdout'] == 'test'].shape[0]
test_predX = testX * c[1] + c[0]
MSE = ((testy - test_predX)**2).sum()/n_test

ax = sns.scatterplot(x='X',y='y',data=df[df['holdout'] == 'test'], label='test data',
                    color='g')
ax.plot(predX,evalX,'k',label='model fit')
ax.text(0.7,0.2,'MSE: {:.2f}'.format(MSE), transform=ax.transAxes,fontsize=15)
ax.legend()
plt.savefig('../figures/holdout_tested.png',dpi=200)
plt.show()
```
